### Finding the Most Popular apartment in the four cities of Fremont, Berkely, Dublin, Pleasanton

```python
style =pd.crosstab(s['city'],s['Apt_Type'])
style 
```

```python
style.plot.bar(figsize=(10,8))
plt.xlabel('City', fontsize=16)
plt.ylabel('Count of Apartment', fontsize=16)
plt.title('Popular House types across different locations',fontsize=20)
plt.legend(title='Apartment Style',loc='upper left')
plt.xticks(rotation=0)
plt.show()

plt.savefig('pht.png',dpi=400)
```

### Finding The Max, MIn and Mean Price of apartments across different locations

```python
comparison = s[['Apt_Type','Price']]
c = comparison.groupby('Apt_Type').mean().round(0)
c = c.reset_index()
c.columns =['Apt_Type','Mean_P']
```

```python
c
```

```python
c1 = comparison.groupby('Apt_Type').min().round(0)
c1 = c1.reset_index()
c1.columns =['Apt_Type','Min_P']
```

```python
c1
```

```python
c2 = comparison.groupby('Apt_Type').max().round(0)
c2 = c2.reset_index()
c2.columns =['Apt_Type','Max_P']
```

```python
c2
```

```python
c3 = comparison.groupby('Apt_Type').median().round(0)
c3 = c3.reset_index()
c3.columns =['Apt_Type','Median_P']
```

```python
c3
```

```python
df2 = pd.merge(c,c1, left_on='Apt_Type', right_on='Apt_Type')
df1 = pd.merge(df2,c2, left_on='Apt_Type', right_on='Apt_Type')
```

```python
df = pd.merge(df1,c3, left_on='Apt_Type', right_on='Apt_Type')
df.set_index('Apt_Type', inplace=True)
df
```

```python
df.plot.bar(figsize=(10,8))
plt.xlabel('Apartment Type', fontsize=16)
plt.ylabel('Price', fontsize=16)
plt.title('Max,Min,Mean,Median price comparison across different locations',fontsize=20)
plt.legend(title='Apartment Style',loc='upper left')
plt.xticks(rotation=0)
plt.show()
```
